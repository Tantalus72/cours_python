
### Introduction

`key=lambda x: x[1]` permet de :

1. Trier des listes de tuples ou de listes
2. Trouver des éléments maximaux ou minimaux selon un critère
3. Manipuler des dictionnaires ou objets complexe


## 1. Trier une liste de tuples en fonction du deuxième élément

# Exemple :

```python
# Liste de tuples (nom, score)
data = [("Alice", 85), ("Bob", 70), ("Charlie", 90)]

# Trier par le score (2ème élément)
sorted_data = sorted(data, key=lambda x: x[1])
print("Données triées par score :", sorted_data)
```

## 2. Trouver le maximum et le minimum dans une liste de tuples

# Exemple :
```python
# Liste de tuples (nom, score)
data = [("Alice", 85), ("Bob", 70), ("Charlie", 90)]

# Trouver le tuple avec le score maximum
max_score = max(data, key=lambda x: x[1])

# Trouver le tuple avec le score minimum
min_score = min(data, key=lambda x: x[1])

print("Tuple avec le score maximum :", max_score)
print("Tuple avec le score minimum :", min_score)
```

## **3. Trier une liste de listes en fonction d’un élément spécifique**

# Exemple :
```python
# Liste de listes [nom, score]
data = [["Alice", 85], ["Bob", 70], ["Charlie", 90]]

# Trier par le score (2ème élément)
sorted_data = sorted(data, key=lambda x: x[1])

print("Données triées par score :", sorted_data)
```


## **4. Trier un dictionnaire par valeurs**

# Exemple :
```python
# Dictionnaire de scores
data = {"Alice": 85, "Bob": 70, "Charlie": 90}

# Trier par les scores (valeurs)
sorted_data = sorted(data.items(), key=lambda x: x[1])

print("Dictionnaire trié par valeurs :", sorted_data)
```

## **5. Trier par plusieurs critères**

# Exemple :
```python
# Liste de tuples (nom, score, âge)
data = [("Alice", 85, 20), ("Bob", 85, 18), ("Charlie", 90, 19)]

# Trier par score (2ème élément), puis par âge (3ème élément)
sorted_data = sorted(data, key=lambda x: (x[1], x[2]))

print("Données triées par score et âge :", sorted_data)
```



## **6. Utiliser avec des objets complexes**

# Exemple :
```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

# Liste d'objets Étudiant
students = [Student("Alice", 85), Student("Bob", 70), Student("Charlie", 90)]

# Trier par l'attribut score
sorted_students = sorted(students, key=lambda x: x.score)

for student in sorted_students:
    print(student.name, student.score)
```


## **7. Trouver l’indice correspondant à une valeur maximale/minimale**

# Exemple :
```python
# Liste de tuples (nom, score)
data = [("Alice", 85), ("Bob", 70), ("Charlie", 90)]

# Trouver l'indice du tuple avec le score maximum
max_index = max(enumerate(data), key=lambda x: x[1][1])[0]

print("Indice du score maximum :", max_index)
```
